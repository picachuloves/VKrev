```python
reviews = pd.read_csv('reviews.csv', sep='\t')
data_pos_reviews = reviews[reviews.sentiment == 'positive']['review'].reset_index(drop=True)
data_neg_reviews = reviews[reviews.sentiment == 'negative']['review'].reset_index(drop=True)
```

```python
data_pos = data_pos_reviews
data_neg = data_neg_reviews
```

```python
size_pos = data_pos.shape[0]
size_neg = data_neg.shape[0]
labels = [1] * size_pos + [0] * size_neg

data_size = size_neg + size_pos
data_pos = data_pos.head(data_size)
data_neg = data_neg.head(data_size)
data = data_pos.append(data_neg, ignore_index=True)
size_pos, size_neg
```

```python
russian_stopwords = stopwords.words("russian")
def preprocess_text(text):
    text = text.lower().replace("ё", "е")
    text = re.sub('[^а-яА-Я]+', ' ', text)
    tokens = text.split(' ')
    tokens = [token for token in tokens if token not in russian_stopwords]
    text = " ".join(tokens)
    return text


datap = [preprocess_text(t) for t in data]
```

```python
x_train, x_test, y_train, y_test = train_test_split(vectors2, labels, test_size=0.3, random_state=1)
y_expect = y_test
```

```python
print('LogReg')

clf_lr = LogisticRegression(random_state=87, max_iter = 500)
parameters = {
#     'class_weight': ['balanced', {1: 0.1, 0: 0.9}, {1: 0.2, 0: 0.8},{1: 0.3, 0: 0.7},{1: 0.4, 0: 0.6}],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
}
# clf = GridSearchCV(clf_lr, parameters)
clf = LogisticRegression(solver='newton-cg', random_state=0)
clf.fit(x_train, y_train)
svm_predict = clf.predict(x_test)

print(cross_val_score(clf, x_train, y_train, cv=5))
print(accuracy_score(y_expect, svm_predict))
print(recall_score(y_expect, svm_predict))
print(precision_score(y_expect, svm_predict))
print(f1_score(y_expect, svm_predict))
```
