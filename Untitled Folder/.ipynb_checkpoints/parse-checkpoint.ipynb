```python
import math
import pandas as pd
import vk
import numpy as np
```

```python
token = '17d81369af7030637ea2b69f310d8a3590225a7093f487da3db6b15386614750c08fd0421274d969878e7'

session = vk.Session(access_token=token)
vk_api = vk.API(session=session)
```

```python
group_id = 64545476
topic_id = 29268766
count = 3748
cel = count // 2000
b = []
drob = math.ceil((count - cel * 2000) / 100)
```

```python
for i in range(cel):
    b.append(20)
b.append(drob)
res = pd.DataFrame()
for i in range(cel + 1):
    inner_count = b[i]
    req = f"""var i = 0;
                      var ans = [];
                      while (i != {inner_count}) {{
                      ans.splice( i * 100, 0, API.board.getComments({{"group_id": {group_id}, "topic_id": {topic_id}, "count": 100, "offset": (i + {i * 20}) * 100}}).items);
                      i = i + 1;
                      }};
                      return ans;
            """
    test = vk_api.execute(code=req, v=5.95)
    for j in range(len(test)):
        res = pd.concat([res, pd.DataFrame(test[j])])
```

```python
res
```

```python
res = res[(res['text'] != '') & (res['from_id'] > 0)].reset_index(drop=True)
res
```

```python
ids = res['from_id']
ids = ids.unique()
ids = ids.tolist()
ids = [str(int) for int in ids]
ids
```

```python
users_count = len(ids)
users_count
```

```python
cel = math.ceil(users_count / 1000)
profiles = pd.DataFrame()
for i in range(cel):
    ids_ = ",".join(ids[i * 1000: (i+1)*1000])
    test = vk_api.users.get(user_ids=ids_, fields='bdate, city,country,occupation,relation,sex', offset=i*1000, v=5.95)
    profiles = pd.concat([profiles, pd.DataFrame(test)])
profiles = profiles.reset_index(drop=True)
```

```python
profiles.shape
```

```python
result = pd.merge(res, profiles, how='left', left_on='from_id', right_on='id')
result
```
